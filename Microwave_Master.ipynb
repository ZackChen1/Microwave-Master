```python
import tkinter as tk
from PIL import Image, ImageTk
from itertools import count, cycle

def button_click():
    time['text'] = f'{round(int(input2.get()) * (int(input1.get())/int(input3.get())))}s'

class ImageLabel(tk.Label):
    """
    A Label that displays images, and plays them if they are gifs
    :im: A PIL Image instance or a string filename
    """
    def load(self, im):
        if isinstance(im, str):
            im = Image.open(im)
        frames = []

        try:
            for i in count(1):
                frames.append(ImageTk.PhotoImage(im.copy()))
                im.seek(i)
        except EOFError:
            pass
        self.frames = cycle(frames)

        try:
            self.delay = im.info['duration']
        except:
            self.delay = 100

        if len(frames) == 1:
            self.config(image=next(self.frames))
        else:
            self.next_frame()

    def unload(self):
        self.config(image=None)
        self.frames = None

    def next_frame(self):
        if self.frames:
            self.config(image=next(self.frames))
            self.after(self.delay, self.next_frame)

#demo :
root = tk.Tk()
root.title('Microwave Master')
root.minsize(width=500, height=300)

lbl = ImageLabel(root)
lbl.pack()
lbl.load('animated.gif')

label = tk.Label(text='How long should I ding?', font=('Apple Chancery', 36, 'bold'))
label.pack()

button = tk.Button(text='Click me senpai', command=button_click)
button.place(x=180, y=400)

input1_label = tk.Label(root, text="Instruction Watts:", font=('Areil', 18, 'bold'))
input1_label.place(x=100, y=80)
input1 = tk.Entry(width=10)
input1.place(x=260, y=80)

input2_label = tk.Label(root, text="Instruction Time:", font=('Areil', 18, 'bold'))
input2_label.place(x=100, y=120)
input2 = tk.Entry(width=10)
input2.place(x=260, y=120)

input3_label = tk.Label(root, text="Microwave Watts:", font=('Areil', 18, 'bold'))
input3_label.place(x=100, y=160)
input3 = tk.Entry(width=10)
input3.place(x=260, y=160)

reccomended_time = tk.Label(root, text="Recommended Time:", font=('Areil', 18, 'bold'))
reccomended_time.place(x=120, y=250)

time = tk.Label(root, text="", font=('Areil', 18, 'bold'))
time.place(x=310, y=250)
root.mainloop()
```
